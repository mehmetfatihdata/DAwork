# <span style=" color:Purple"> Catplot</span>

```python
sns.catplot(x='species', y = "bill_length_mm",data=penguins)
```

```python
sns.catplot(x='species', y = "bill_length_mm",data=penguins ,kind="bar",hue="sex" )
```

```python
sns.catplot(x='species', y = "bill_length_mm",data=penguins ,kind="bar",hue="sex", col="island" )
```

```python
sns.catplot(x='species', y="body_mass_g", data=penguins, kind="bar", row="sex", col="island" )
```

```python
sns.catplot(x='species', y="body_mass_g", data=penguins, kind="box", row="sex", col="island" )
```

# <span style=" color:Purple"> Jointplot </span>

```python
sns.jointplot(x='flipper_length_mm',y='body_mass_g',data=penguins)  # kind : { "scatter" | "kde" | "hist" | "hex" | "reg" | "resid" }
```

```python
sns.jointplot(x='flipper_length_mm',y='body_mass_g',data=penguins, kind="reg")
```

```python
sns.jointplot(x='flipper_length_mm',y='body_mass_g',data=penguins, hue="species")
```

# <span style=" color:Purple"> Pairplot </span>

```python
sns.pairplot(penguins ,hue = 'species',diag_kind = "kde",kind = "scatter",palette = "husl", markers=["o", "s", "D"]);
```

```python
############################################################################################################################
```

<div class="alert alert-block alert-success ">

# <span style=" color:Purple"> Pairplot </span>

```python
g = sns.pairplot(penguins, diag_kind="kde")
g.map_lower(sns.kdeplot, levels=4, color=".2")
plt.suptitle('Pair Plot of Penguins', 
             size = 28, c ='r')
plt.tight_layout();
```

```python
############################################################################################################################
```

```python
g = sns.FacetGrid(data = penguins, row = "species", col =  "island", hue = "sex")
g = g.map(sns.kdeplot, "body_mass_g")
```
